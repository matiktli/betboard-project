```python
import pandas as pd

pl_18_19 = pd.read_csv('../resources/PremierLeague_18_19.csv')
pl_19_20 = pd.read_csv('../resources/PremierLeague_19_20.csv')
cl_18_19 = pd.read_csv('../resources/Championship_18_19.csv')
cl_19_20 = pd.read_csv('../resources/Championship_19_20.csv')

data = [pl_18_19,pl_19_20,cl_18_19,cl_19_20]
data = pd.concat(data)
```

```python
import matplotlib.pyplot as plt

data = data.head(20)
fig = plt.figure()

plt.plot(data['Date'], data['FTHG'])
plt.plot(data['Date'], data['FTAG'])
plt.show()
```
